```python
import cv2
from matplotlib import pyplot as plt
from numpy import linalg
import numpy as np
```

```python
image_bgr = cv2.imread('sample-images/toe-socks.jpg')
image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
R0, G0, B0 = image[:, :, 0], image[:, :, 1], image[:, :, 2]
zeros = np.zeros_like(R0)
```

```python
image_R0 = image.copy() ; image_R0[:, :, 1:] = 0
image_G0 = np.stack([zeros, G0, zeros], axis=-1)
image_B0 = image.copy() ; image_B0[:, :, :2] = 0

fig = plt.figure(figsize=(15, 10))

fig.add_subplot(2, 2, 1)

plt.imshow(image_R0)
plt.axis('off')
plt.title('Red')

fig.add_subplot(2, 2, 2)

plt.imshow(image_G0)
plt.axis('off')
plt.title('Green')

fig.add_subplot(2, 2, 3)

plt.imshow(image_B0)
plt.axis('off')
plt.title('Blue')

fig.add_subplot(2, 2, 4)

plt.imshow(image)
plt.axis('off')
plt.title('Original')
```
